```python
import pandas as pd
c=pd.read_csv('train.csv')
e=c.iloc[:,-1]
print(c.columns)
c.isna().sum()
c 
```

```python
ytrain=c.iloc[:,0].values
xtrain=c.iloc[:,1:-1].values
len(xtrain[0])
```

```python
from sklearn.preprocessing import LabelEncoder
l=LabelEncoder()
ytrain=l.fit_transform(ytrain)
ytrain
```

```python
tes=pd.read_csv('test.csv')
xtest=tes.iloc[:,:-1].values
xtest
ids=tes.iloc[:,-1].values
list(ids)
```

```python
from sklearn.ensemble import RandomForestClassifier
rf=RandomForestClassifier()
rf.fit(xtrain,ytrain)
```

```python
ypred=rf.predict(xtest)
len(list(ypred))
```

```python
import csv
for i in range(2500):
    a=[]
    a.append(ids[i])
    a2=list(ypred[i])
    a.append(''.join(a2))
    with open('sample_submission.csv','a',newline='') as csv1:
        writer = csv.writer(csv1)
        writer.writerow(a)
        csv1.close()
```

```python
from sklearn.preprocessing import LabelEncoder,OneHotEncoder
o=OneHotEncoder(categorical_features=[0])
ytrain=o.fit_transform(ytrain).toarray()
ytrain
```

```python
import keras
from keras.models import Sequential
from keras.layers import Dense
from keras.models import Sequential, load_model
from keras.layers.core import Dense, Dropout, Activation
from keras.utils import np_utils
```

```python
model = Sequential()
model.add(Dense(6, input_shape=(10,)))
model.add(Activation('relu'))                            
model.add(Dense(6))
model.add(Activation('relu'))
model.add(Dense(6))
model.add(Activation('relu'))
model.add(Dense(4))
model.add(Activation('softmax'))
```

```python
model.compile(optimizer='adam',loss='categorical_crossentropy',metrics=['accuracy'])
model.fit(xtrain,ytrain,batch_size=200,nb_epoch=100)
```
